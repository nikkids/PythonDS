# **Regular Expression Combined with Pandas**

```python
import re
#Menggunakan regular expression
#akhir nama (Gunakan $)
genshin.loc[genshin['character_name'].str.contains(r'li$'),['character_name']]
#awal name (Gunakan ^)
genshin.loc[genshin['character_name'].str.contains(r'^[G]'),['character_name']]

#bisa menggunakan startswith
genshin.loc[genshin['region'].isin(['Mondstadt','Sumeru']) & (genshin.character_name.str.startswith("K")),['character_name']]
```

```python
import re
genshin.loc[10:40, 'character_name'][~genshin.loc[10:40, 'character_name'].str.startswith('K',re.IGNORECASE)]
#bisa endswith juga
#ini mencari rentang 10 hingga 40 tanpa nama yang berawalan k
```

```python
#Mengembalikan karakter awalan K dan akhiran A dimana memiliki atk_1_20 lebih dari 19 dan berasal dari Inazuma
result = genshin.loc[
    (genshin['atk_1_20'] >= 19) &
    (genshin['character_name'].str.contains(r'^K.*a$', re.IGNORECASE)) &
    (genshin['region'] == 'Inazuma'),['character_name','atk_1_20','region']
]
result

#add 2 conditional in loc
# ^K = Awalan K
# .* = karakter apa saja (bagus untuk ignore)
# s$ = Akhiran S
```

```python
result.to_csv('result.csv')
```

# **Mengkategorikan Data**

```python
genshin['atk_90_90'].describe()
```

```python
#Menggunakan Lambda
genshin['Atk_Tier_List'] = genshin['atk_90_90'].apply(lambda x: 'Bad' if x < 200 else ('Mid' if x < 300 else 'OP'))

bad = genshin.loc[genshin.Atk_Tier_List == 'Bad', ['character_name']].count()
mid = genshin.loc[genshin.Atk_Tier_List == 'Mid', ['character_name']].count()
OP = genshin.loc[genshin.Atk_Tier_List == 'OP', ['character_name']].count()
print(bad['character_name'],mid['character_name'],OP['character_name'])
```

```python
#Menggunakan Function
def kategorirarity(baris):
  if baris['star_rarity'] == 4 and baris['region'] == 'Mondstadt':
      return 'Mondstadt 4 Star'
  elif baris['star_rarity'] == 4 and baris['region'] == 'Liyue':
      return 'Liyue 4 Star'
  elif baris['star_rarity'] == 4 and baris['region'] == 'Inazuma':
      return 'Inazuma 4 Star'
  elif baris['star_rarity'] == 4 and baris['region'] == 'Sumeru':
      return 'Sumeru 4 Star'
  elif baris['star_rarity'] == 4 and baris['region'] == 'Fontaine':
      return 'Fontaine 4 Star'
  else:
      return '5 Star'

genshin['Rarity'] = genshin.apply(kategorirarity, axis = 1)
#axis 1 berarti row, axis 0 berarti kolom
genshin.loc[genshin.Rarity == 'Liyue 4 Star', ['character_name']].count()
```
