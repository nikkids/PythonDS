# **Bar Plot**

```python
program = ['C++','python','javascript','lua','go']
vote = [50,150,30,10,70]
```

```python
plt.bar(program,vote, color = 'red', align = 'edge', width = 0.4, edgecolor = 'black', lw = 5)
#align untuk atur lokasi tulisan dibawahnya
#width untuk mengatur lebar plotnya
#edgecolor untuk nambah tepi-tepinya
#lw untuk ketebalan edgecolor
plt.show()
```

# **Histogram**

```python
distribusi_normal = np.random.normal(20,1.5,100)
#.normal(x,y,z)
# x = mean
# y = standar deviasi
# z = banyak data
# standar deviasi kecil = 0.7 (contoh)
# standar deviasi besar = 8.6 (contoh)
# SD kecil berarti data rata-rata seragam, SD besar berarti data tidak seragam atau terdapat outlier
plt.hist(distribusi_normal)
plt.show()
```

```python
plt.hist(distribusi_normal, bins = 5)
#bins untuk menentukan mau berapa kotak (dicontoh ini jadi 5 kotak doang)
plt.hist(distribusi_normal,
         bins = [distribusi_normal.min(),19,20,21, distribusi_normal.max()])
#kalau ini dibagi bagi jadi ada box yang min hingga max, ada juga box yang 19,20,21
```

```python
#Cumulative Historgram
plt.hist(distribusi_normal, bins = 20, cumulative = True)

#Bar tiap datanya adalah jumlah dari data sebelumnya
#Contoh bar ke 24 (24+23+22+21+20)
#Contoh bar ke 17 (17+16)
```

# **Subplot**

```python
x = np.arange(100)
#generate angka 1-100

style.use('Solarize_Light2')

fig,axis = plt.subplots(2,2)

axis[0,0].plot(x, np.sin(x))
axis[0,0].set_title('sin')

axis[0,1].plot(x, np.cos(x))
axis[0,1].set_title('cos')

axis[1,0].plot(x, np.random.random(100))
axis[1,0].set_title('random')

axis[1,1].plot(x, np.log(x))
axis[1,1].set_title('log')

fig.suptitle('Subplot')
plt.tight_layout()

plt.savefig('Subplot2.png', dpi =300)
plt.show()
```

# **3D Plot**

```python
ax = plt.axes(projection = '3d')

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x,y,z)
ax.set_title('3D')
plt.show()
```

```python
ax = plt.axes(projection = '3d')

x = np.arange(0,50, 0.2)
y = np.cos(x)
z = np.tan(x)

ax.plot(x,y,z)
ax.set_title('3D')
plt.show()
```

```python
ax = plt.axes(projection = '3d')
x = np.arange(1,10,0.1)
y = np.arange(1,10,0.1)

X,Y = np.meshgrid(x,y)
Z = np.sin(X) + np.sin(Y)
ax.plot_surface(X,Y,Z, cmap = 'Spectral')
plt.show()
```
